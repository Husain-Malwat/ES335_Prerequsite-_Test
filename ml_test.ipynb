### Matrix Multiplication:

When you multiply a matrix A with dimensions  $ n \times k$  by another matrix B with dimensions $k \times m$, the resulting matrix C (with dimensions $n \times m$) is obtained by taking the dot product of each row of A with each column of B.

#### Number of Multiplications:

1. **Basic Idea:**
   - In a matrix multiplication of A (with dimensions \(n \times k\)) and B (with dimensions \(k \times m\)), each entry in the result matrix C (with dimensions $n \times m$) is calculated by taking the dot product of a row from A and a column from B.

2. **Element Calculation:**
   - Let's consider an arbitrary entry in the result matrix C at position (i, j), denoted as $C_{ij}$.
   - This element $C_{ij}$ is calculated as the sum of the products of corresponding elements from the $i$th row of A and the $j$th column of B:

     $ C_{ij} = \sum_{l=1}^{k} A_{il} \cdot B_{lj} $

3. **Multiplications per Entry:**
   - In the summation above, each term $A_{il} \cdot B_{lj}$ involves a multiplication operation.
   - There are $k$ terms in the summation (from $l=1$ to $l=k$), so there are $k$ multiplications for each entry in the result matrix.

4. **Total Multiplications:**
   - Since there are $n \times m$ entries in the result matrix (n rows and m columns), and each entry involves $k$ multiplications, the total number of multiplications in the matrix multiplication is given by:
     $ \text{Total Multiplications} = n \times m \times k $



#### Number of Additions:

For each element in the result matrix, there are $k-1$ additions because we are summing up $k$ products. Therefore, the total number of additions in the matrix multiplication is $ n \times m \times (k-1) $.

#### Summary:

In summary, the matrix multiplication of A and B involves $ n \times m \times k $ multiplications and $ n \times m \times (k-1) $ additions. These calculations help us understand the computational complexity and the number of operations involved in matrix multiplication.

```python
import timeit
import statistics as st
import random
import numpy as np


n = int(input("Enter the number of rows of the first matrix: "))
k = int(input("Enter the number of columns of the first matrix: "))
m = int(input("Enter the number of columns of the second matrix: "))

```

```python
timetaken_list = []
timetaken_np = []

for i in range(100):
    
    int_min = -2147483647
    int_max = 2147483647
    # Generate random matrices matrix1 and matrix2 with elements between INT_MIN and INT_MAX
    matrix1 = [[random.randint(int_min, int_max) for _ in range(k)] for _ in range(n)]
    matrix2 = [[random.randint(int_min, int_max) for _ in range(m)] for _ in range(k)] 
    result = [[0 for _ in range(m)] for _ in range(n)]

    begin = timeit.default_timer()
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for h in range(len(matrix2)):
                result[i][j] += matrix1[i][h] * matrix2[h][j]
    end = timeit.default_timer()
    duration = round((end - begin) * 10 ** 6, 3)  
    timetaken_list.append(duration)

    

    # ----------------------------------------------------------------------------------------------------------------

    np_matrix1 = np.array(matrix1)
    np_matrix2 = np.array(matrix2)

    begin_np = timeit.default_timer()
    result_np = np.dot(matrix1, matrix2)
    end_np = timeit.default_timer()

    duration_np = round((end_np - begin_np) * 10 ** 6, 3) 
    timetaken_np.append(duration_np)
    
mean_list = st.mean(timetaken_list)
print("Average time taken for list implementation: ", mean_list)
mean_list_np = st.mean(timetaken_np)    
print("Average time taken for numpy implementation: ", mean_list_np)
```

## Numpy is almost 100 times faster than list based matrix multiplication approach, this can be attributed to several factors:

### Algorithmic Differences

NumPy uses highly optimized algorithms, often leveraging BLAS (Basic Linear Algebra Subprograms) libraries, for matrix operations. These libraries are implemented in lower-level languages like Fortran or C and are tuned for maximum performance.

### Memory Access Patterns

NumPy efficiently manages memory access patterns, which is crucial for performance. The C implementation shown uses nested loops with direct memory access, but it may not be optimized for cache efficiency or aligned memory access, which can impact performance.

### Vectorization

NumPy is designed to take advantage of vectorization instructions provided by modern CPUs. This means that operations can be performed on multiple data elements simultaneously, leading to significant speedups. The C implementation may not be leveraging such optimizations.

### Compiler Optimization

NumPy operations are often implemented in compiled languages with aggressive compiler optimizations. The C code provided may not be fully optimized by the compiler, depending on the compilation flags used.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def func1(x):
    return x

def func2(x):
    return x**2

def func3(x):
    return (x**3) / 100

def func4(x):
    return np.sin(x)

def func5(x):
    return np.sin(x) / x

def func6(x):
    return np.log(x)

def func7(x):
    return np.exp(x)

# Create subplots
fig, axs = plt.subplots(4, 2, figsize=(10, 14))

# Plot each function
functions = [func1, func2, func3, func4, func5, func6, func7]
titles = ['$y=x$', '$y=x^2$', '$y=x^3/100$', '$y=sin(x)$', '$y=sin(x)/x$', '$y=log(x)$', '$y=e^x$']

# Iterate through subplots in the first three rows
for i, ax in enumerate(axs[:-1, :].flatten()):
    x = np.linspace(0.5, 100.0, 200)
    y = functions[i+1](x)
    ax.plot(x, y, label=titles[i])
    ax.set_title(f'{i}. {titles[i]}')
    ax.legend(loc='upper left')
    ax.grid()

# Plot the last subplot in the last row
x = np.linspace(0.5, 100.0, 200)
y = functions[-1](x)
axs[-1, 0].plot(x, y, label=titles[-1])
axs[-1, 0].set_title(f'{titles[-1]}')
axs[-1, 0].legend(loc='upper left')
axs[-1, 0].grid()

# Remove empty subplots
for i in range(1, 2):
    axs[-1, i].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
